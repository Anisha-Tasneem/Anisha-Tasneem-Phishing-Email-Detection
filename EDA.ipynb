```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (replace 'your_file_path.csv' with the actual file path)
df = pd.read_csv('/content/drive/MyDrive/CSE445/Phishing_Email.csv')

email_type = df['Email Type'].value_counts()  # Use 'df' instead of 'loaded_dataset'

# Pie chart
plt.pie(email_type, colors=['yellow', 'red'], autopct='%1.1f%%')
plt.title('Total of email texts')
plt.legend(['Safe', 'Phishing'], title='Email Type')
plt.show()
```
